```python
import pandas as pd
```

```python
df =pd.read_csv("./kc_house_data.csv")
```

```python
df.head()
```

```python
adf=df[['price','bedrooms','sqft_living','waterfront']].copy()
adf.head()
adf.sort_values(by=['price'])
```

```python
for i in range(1,int(adf.price.max()/100000)+1):

    adf['binned_prices']= adf.apply(lambda x: '['+str((i-1)*100000)+'-'+ str(i*100000)+']' if ((i-1)*100000 <= x['price'] < i*100000 ) else x['binned_prices'], axis=1)
```

```python
for i in range(1,int(adf.price.max()/100000)):

    adf['binned_calc']= adf.apply(lambda x: i*100000 if ((i-1)*100000 <= x['price'] < i*100000 ) else x['binned_calc'], axis=1)
```

```python
adf.info()
```

```python
z= adf[['binned_calc','binned_prices','bedrooms','sqft_living','waterfront']].groupby(['binned_calc','binned_prices']).mean().round(1)
z
```

```python
z.corr()
```

We can see the expected trend of a positive strong correlation between the price ranges, number of bedrooms and living area. However we have a weaker correlation with the waterfront suggesting that higher prices might not necessarily mean the property will be located at the waterfront.
